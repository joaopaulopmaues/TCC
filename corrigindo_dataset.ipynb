```python
i=0
out=[[]]*2240
for el in laux:
    if not ((el[0] in valid_calls) and (el[499] in valid_calls)) and el[0]!='A' and el[499]!='A':
        out[i]=el[:500]+el[501:]+['A']
        #print('INVALID CALL: ',len(el),i,el[0] in valid_calls,el[499] in valid_calls, el[0], el[499],el[500],el[501],el[509],el[510],el[519])
        #print('len(el)=',len(el),el[-21:]) #tudo ok, não há mutex em nenhuma das linhas nem dlls fora do lugar
        i+=1
        continue
    elif len(el)<520:
        out[i]=el+['A'] #não há problemas com relacao a mutex ou dlls fora do lugar nem antes nem após inserção ou remoção
        #print("SHORT LINE: ",len(el),i,'el[0]:',el[0],'el[499]',el[499],'el[500]',el[500],'el[509]',el[509],'el[518]',el[518])
        i+=1
        continue
    elif len(el)==520 and el[500]=='A':
        if (el[511] in valid_dlls or el[511]=='apimswin' or el[511].endswith('dll') or el[511].endswith('DLL')):
            mutex=[]
            for dll in el[511:]:
                if not dll in valid_dlls and dll != 'apimswin':
                    mutex.append(dll)
            while len(mutex)<10:
                mutex.append('A')
            out[i]=el[:500]+el[501:511]+mutex #está certinho, testado
            i+=1
            continue
            #print('DEMAIS CASOS: ',len(el),el[509],el[510],el[511],el[519])
        elif (el[509] in valid_dlls and (el[510] in valid_dlls or el[510]=='A')) or (el[509]=='A' and el[510]=='A' and el[511]=='A' and el[519]=='A'):
            out[i]=el[:500]+el[501:]+['A']
            #print(el[510:]) #sem problemas
            i+=1
            continue
        elif el[510]=='A':
            out[i]=el[:500]+el[501:]+['A']
            #print(el[510:]) #sem problemas
            i+=1
            continue
    elif not ((el[511] in valid_mutexes) and (el[520] in valid_mutexes)) and el[511]!='A' and el[520]!='A':
        mauxl=[]
        for mutex in el[-10:]:
            if mutex in valid_mutexes:
                mauxl.append(mutex)
        while len(mauxl)<10:
            mauxl.append('A')
        out[i]=el[:500]+el[501:511]+mauxl
        #print('INVALID MUTEX: ',el[498:500]+el[501:511]+mauxl)
        i+=1
        continue
    i+=1
```
